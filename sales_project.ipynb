```python
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
data ={ 
    'Month': ['Jan','Feb','Mar','Apr','May','Jun',
              'Jan','Feb','Mar','Apr','May','Jun',
              'Jan','Feb','Mar','Apr','May','Jun'],
    'Region': ['Dubai','Dubai','Dubai','Dubai','Dubai','Dubai',
               'Abu Dhabi','Abu Dhabi','Abu Dhabi','Abu Dhabi','Abu Dhabi','Abu Dhabi',
               'Sharjah','Sharjah','Sharjah','Sharjah','Sharjah','Sharjah'],
    'Product': ['Electronics','Furniture','Electronics','Furniture','Electronics','Furniture',
                'Electronics','Furniture','Electronics','Furniture','Electronics','Furniture',
                'Electronics','Furniture','Electronics','Furniture','Electronics','Furniture'],
    'Sales': [15000,8000,17000,9000,20000,11000,
              12000,6000,13000,7000,15000,8000,
              8000,4000,9000,5000,10000,6000],
    'Profit': [3000,1200,3500,1500,4000,1800,
               2400,900,2600,1100,3000,1300,
               1600,600,1800,800,2000,1000],
    'Discount': [0.1,0.2,0.1,0.15,0.05,0.2,
                 0.1,0.2,0.1,0.15,0.05,0.2,
                 0.1,0.2,0.1,0.15,0.05,0.2]
}
    
df = pd.DataFrame(data) 
print("Data created successfully"  )
print(df.shape)
print(df.head())
```

```python
print("=== REGIONAL SALES PERFORMANCE ===")
regional =df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print(regional) 

print("\n=== REGIONAL PROFIT === ")
regional_profit = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
print(regional_profit)
```

```python
print ("=== PRODUCT PERFORMANCE ===")
product_performance = df.groupby('Product')['Sales'].sum().sort_values(ascending=False)
print(product_performance)

print("\n=== PRODUCT PROFIT ===")
product_profit = df.groupby('Product')['Profit'].sum().sort_values(ascending=False)
print(product_profit)
```

```python
print("=== MONTHLYT SALES TREND === ")
monthly_sales = df.groupby('Month')['Sales'].sum()
print(monthly_sales)

monthly_sales.plot(kind='line', marker='o', color='blue', title='Monthly Sales Trend UAE')
plt.xlabel('Month')
plt.ylabel('Sales (AED)')
plt.tight_layout()
plt.show()
```

```python
regional =df.groupby('Region')['Sales'].sum()
regional.plot(kind='bar', color=['blue', 'orange', 'green'], title='Sales by Region UAE')
plt.xlabel('Region') 
plt.ylabel('Sales (AED)')
plt.tight_layout()
plt.show()
```
