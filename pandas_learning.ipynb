```python
import pandas as pd 

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Region': ['North', 'South', 'East', 'West', 'North'],
    'sales': [100, 150, 200, 250, 300],
    'profit': [20, 30, 40, 50, 60]
}
df = pd.DataFrame(data)
print(df)
```

```python
print(df.shape)
```

```python
print(df.describe())
```

```python
print("total sales:",df['sales'].sum()) 
```

```python
print("Average profit:", df['profit'].mean())
```

```python
print("highest sales person:")
print(df[df['sales'] == df['sales'].max()])
```

```python
dubai =df[df['Region'] == 'North']
print(dubai)
```

```python
high = df[df['sales'] > 150]
print(high)
```

```python
sorted_df = df.sort_values('sales', ascending=False)
print(sorted_df)
```
