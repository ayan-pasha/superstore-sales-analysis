```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(':memory:')

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Region': ['North', 'South', 'East', 'West', 'North'],
    'sales': [100, 150, 200, 250, 300],
    'profit': [20, 30, 40, 50, 60]
}

df = pd.DataFrame(data)
df.to_sql('sales_data', conn, index=False)

# Query 1 - Select specific columns
print("=== Query 1 - Specific Columns ===")
result = pd.read_sql_query("SELECT Name, sales FROM sales_data", conn)
print(result)

# Query 2 - WHERE filter
print("=== Query 2 - Sales above 150 ===")
result = pd.read_sql_query("SELECT * FROM sales_data WHERE sales > 150", conn)
print(result)

# Query 3 - ORDER BY
print("=== Query 3 - Sorted by Sales ===")
result = pd.read_sql_query("SELECT * FROM sales_data ORDER BY sales DESC", conn)
print(result)

# Query 4 - GROUP BY
print("=== Query 4 - Sales by Region ===")
result = pd.read_sql_query("SELECT Region, SUM(sales) as Total_Sales FROM sales_data GROUP BY Region", conn)
print(result)
```

```python
import sqlite3
import pandas as pd
conn=sqlite3.connect(':memory:')

costomers={
    'CustomerID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Region': ['North', 'South', 'East', 'West', 'North']
}

Orders = {
    'OrderID': [101, 102, 103, 104, 105],
    'CustomerID': [1, 2, 3, 4, 5],
    'Sales': [250, 150, 300, 200, 350],
    'Profit': [50, 30, 70, 40, 80]
    
}

df1=pd.DataFrame(costomers)
df2=pd.DataFrame(Orders)

df1.to_sql('customers', conn, index=False)
df2.to_sql('orders', conn, index=False)

query = """
SELECT customers.Name, customers.Region, orders.Sales, orders.Profit
FROM customers
INNER JOIN orders
ON customers.CustomerID = orders.CustomerID
"""
result = pd.read_sql_query(query, conn)
print(result)
```
